```python
from google.colab import drive
drive.mount('/content/drive')

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/datasetD1.csv')
#data2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Loan_defaultD3.csv')
#print(data1)
#print("****************************")

# Seleccionar las dos primeras columnas num√©ricas
X1 = data1.iloc[:, [3, 4]].fillna(data1.iloc[:, [3, 4]].mean()).values
print(X1)
print("****************************")


# Normalizar los datos
scaler = StandardScaler()
X1_normalized = scaler.fit_transform(X1)
print(X1_normalized)
```

```python
# Visualizar datos originales (antes de K-Means)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X1[:, 0], X1[:, 1], s=50, alpha=0.5)
plt.xlabel("$x_1$", fontsize=14)
plt.ylabel("$x_2$", fontsize=14, rotation=0)
plt.title("Datos originales datasetD1")
plt.tight_layout()
plt.show()
```

```python
# Establecer hiperpar√°metros y aplicar K-Means con 20 clusters
kmeans1 = KMeans(n_clusters=20, random_state=42)
y1 = kmeans1.fit_predict(X1_normalized)

# Visualizar clusters
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(X1_normalized[:, 0], X1_normalized[:, 1], c=y1, s=50, cmap='viridis')
plt.xlabel("$x_1$", fontsize=14)
plt.ylabel("$x_2$", fontsize=14, rotation=0)
plt.title("Clusters datasetD1 (20 clusters)")
plt.colorbar(label='Cluster')
plt.tight_layout()
plt.show()
```

```python
# Calcular Silhouette Score para determinar n√∫mero √≥ptimo de clusters
range_n_clusters = range(2, 21)
silhouette_scores1 = []

for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    y1_temp = kmeans.fit_predict(X1_normalized)
    silhouette_scores1.append(silhouette_score(X1_normalized, y1_temp))

# Graficar Silhouette Scores
plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, silhouette_scores1, marker='o', label='datasetD1')
plt.xlabel("N√∫mero de clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score vs N√∫mero de Clusters")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Mostrar el n√∫mero √≥ptimo de clusters (m√°ximo Silhouette Score)
optimal_clusters1 = range_n_clusters[np.argmax(silhouette_scores1)]
print(f"N√∫mero √≥ptimo de clusters para datasetD1: {optimal_clusters1}")

# Mostrar Silhouette Score para 20 clusters
print(f"Silhouette Score para datasetD1 con 20 clusters: {silhouette_score(X1_normalized, y1):.3f}")
```
